```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error,r2_score
```

```python
import pandas as pd
import numpy as np
df= pd.read_csv("sample_data/california_housing_test.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
y=df["median_house_value"]
x=df.drop(["median_house_value"],axis=1)
```

```python
X_train, X_test, y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_lr_train_pred=lr.predict(X_train)
y_lr_test_pred=lr.predict(X_test)

```

```python
print(y_lr_train_pred,y_lr_test_pred)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

MODEL EVAULATION

```python
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
```

```python
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results=pd.DataFrame(data=["Linear Regression",lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=["method","train_mse","train_r2","test_mse","test_r2"]
```

```python
accuracy = accuracy_score(y_train,y_train)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
